1. Implement a solver for the shallow water equations,
$$ \begin{bmatrix} h \\ h u \end{bmatrix}_t + \begin{bmatrix} hu \\ hu^2 + \frac g 2 h^2 \end{bmatrix}_x = 0 $$
where $h$ is water thickness, $hu$ is the momentum of a column of water, and $g$ is the gravitational potential.
The shallow water equations are very similar to isothermal gas dynamics except that the wave speed $c$ is not constant, but rather $\pm \sqrt{gh}$ as can be seen from the eigenvalues of the flux Jacobian.
* Choose initial conditions for your numerical experiments that create all possible configurations of subsonic and supersonic shocks and rarefactions.
* The total energy in the system is kinetic plus potential energy
$$ E = \int_{\Omega} \frac h 2 u^2 + \frac g 2 h^2 . $$
Compare the evolution of total energy using a first order method and a method using slope reconstruction, for a configuration with a shock and a configuration with only rarefactions.
* Does the result depend on your choice of Riemann solver, e.g., between HLL and Rusanov or between an exact solver and HLL?  Does it depend on your choice of limiter in slope reconstruction?

First, some preliminary definitions from class:

```python
%matplotlib inline
import numpy
from matplotlib import pyplot
pyplot.style.use('ggplot')
%run fdtools.py

limiters = []
def limappend(lim):
    limiters.append(lim)
    return lim

@limappend
def limit_zero(r):
    return 0*r

@limappend
def limit_none(r):
    return 0*r+1

@limappend
def limit_minmod(r):
    return numpy.maximum(numpy.minimum(2*r, 2*(1-r)), 0)

@limappend
def limit_sin(r):
    return numpy.where(numpy.all((0 < r, r < 1), axis=0),
                      numpy.sin(numpy.pi*r),
                      0)

@limappend
def limit_vl(r):
    return numpy.maximum(4*r*(1-r), 0)

@limappend
def limit_bj(r):
    return numpy.clip(numpy.minimum(4*r, 4*(1-r)), 0, 1)

r = numpy.linspace(-.1, 1.1, 51)
for lim in limiters:
    pyplot.plot(r, lim(r), label=lim.__name__)
pyplot.legend(loc='lower center');
```

### 1: implement a solver for the shallow water equations
Next, we define the flux function and some approximate Reimann solvers for the shallow water equation:

```python
%matplotlib inline
import numpy
from matplotlib import pyplot
pyplot.style.use('ggplot')
%run fdtools.py
def flux_shallowWater(U, g):
    h = U[0]
    u = U[1] / h
    return numpy.array([U[1], U[1]*u + (g/2)*h**2])

def reimann_hll_shallowWater(UL, UR,g, verbose = False):
    hL = UL[0]
    hR = UR[0]
    uL = UL[1] / hL
    uR = UR[1] / hR
    sL = numpy.minimum(uL - (g*hL)**0.5, uR - (g*hR)**0.5)
    
    if (sL[0] > 0) and sL[0] > ((g*hL)**0.5).any() and verbose:
        print('supersonic shock')
    elif sL[0] > 0 and verbose:
        print('subsonic shock')
        
    sR = numpy.maximum(uL + (g*hL)**0.5, uR + (g*hR)**0.5)
    
    if (sR[0] < 0 ) and numpy.abs(sR[0]) > ((g*hR)**0.5).any() and verbose:
        print('supersonic rarefaction')
    elif sR[0] < 0 and verbose:
        print('subsonic rarefaction')
    fL = flux_shallowWater(UL, g)
    fR = flux_shallowWater(UR, g)
    return numpy.where(sL > 0, fL,
                       numpy.where(sR < 0, fR,
                                   (sR*fL - sL*fR + sL*sR*(UR - UL)) / (sR-sL)))

def reimann_Rusonov_shallowWater(UL, UR,g):
    hL = UL[0]
    hR = UR[0]
    uL = UL[1] / hL
    uR = UR[1] / hR
    sR = numpy.maximum(uL + (g*hL)**0.5, uR + (g*hR)**0.5)
    sL = - sR
    fL = flux_shallowWater(UL, g)
    fR = flux_shallowWater(UR, g)
    return numpy.where(sL > 0, fL,
                       numpy.where(sR < 0, fR,
                                   (sR*fL - sL*fR + sL*sR*(UR - UL)) / (sR-sL)))

def fvsolve2system(riemann, U0, a=-1, b=1, n=20, tfinal=1, limit=limit_minmod, verbose = False, args = ()):
    h = (b - a)/n
    x = numpy.linspace(a+h/2, b-h/2, n) # Element midpoints (centroids)
    U0x = U0(x)
    Ushape = U0x.shape
    idxL = numpy.arange(-1, n-1)
    idxR = numpy.arange(1, n+1) % n
    def rhs(t, U):
        U = U.reshape(Ushape)
        jump = U[:,idxR] - U[:,idxL]
        r = numpy.zeros_like(jump)
        numpy.divide(U - U[:,idxL], jump, out=r, where=(jump!=0))
        g = limit(r) * jump / (2*h) #apply slope limiter
        fluxL = riemann(U[:,idxL] + g[:,idxL] * h/2, U - g * h/2, *args, verbose)
        return (fluxL - fluxL[:,idxR]).flatten() / h
    hist = ode_rkexplicit(rhs, U0x.flatten(), h=h/5, tfinal=tfinal)
    return x, [(t, U.reshape(Ushape)) for t, U in hist]

```

Next, test the solver to make sure it works:

```python
def initial_shallowWater1(x):
    return numpy.array([1 + 2*(numpy.exp(-(x*4)**2)>.5),
                        0*x])

x, hist = fvsolve2system(reimann_hll_shallowWater, initial_shallowWater1, n=200, limit=limit_vl, args=(1,))
for t, U in hist[::len(hist)//5]:
    pyplot.plot(x, U[0], label='t={:.3f}'.format(t))
pyplot.ylabel('height')
pyplot.legend(loc='lower right');
pyplot.figure()
for t, U in hist[::len(hist)//5]:
    pyplot.plot(x, U[1]/U[0], label='t={:.3f}'.format(t))
pyplot.ylabel('velocity')
pyplot.legend(loc='lower right');
```

### 2:  Choose initial conditions that create all possible configurations of subsonic and supersonic shocks and rarefactions.

Subsonic shock: $sL < \sqrt{gh}, sL > 0$

Supersonic shock: $sL > \sqrt{gh}, sL > 0$

Subsonic rarefaction: $|sR| < \sqrt{gh}, sR < 0$

Supersonic rarefaction: $|sR| > \sqrt{gh}, sR < 0$

```python
def initial_shallowWater2(x): #make subsonic shocks
    return numpy.array([1+0*x,
                         1+2*numpy.sin(2*numpy.pi*x)])

def initial_shallowWater3(x): #make supersonic shocks
    return numpy.array([1+0*x,
                         2+2*numpy.sin(2*numpy.pi*x)])

def initial_shallowWater4(x): #make subsonic rarefactions
    return numpy.array([1+0*x,
                         -1-numpy.sin(2*numpy.pi*x)])

def initial_shallowWater5(x): #make supersonic rarefactions
    return numpy.array([1+0*x,
                         -1-2*numpy.sin(2*numpy.pi*x)])

def plotWithIC(IC, verbose = False):
    x, hist = fvsolve2system(reimann_hll_shallowWater, IC, n=200, limit=limit_vl, verbose = verbose, args=(1,))
    for t, U in hist[::len(hist)//5]:
        pyplot.plot(x, U[0], label='t={:.1f}'.format(t))
    pyplot.ylabel('height')
    pyplot.legend(loc='lower right');
    pyplot.figure()
    for t, U in hist[::len(hist)//5]:
        pyplot.plot(x, U[1]/U[0], label='t={:.1f}'.format(t))
    pyplot.ylabel('velocity')
    pyplot.legend(loc='lower right');
    
plotWithIC(initial_shallowWater5, verbose = False)
```

### 3:  Show evolution of energy in system with and without slope reconstruction for a shock and a rarefaction.
$$ E = \int_{\Omega} \frac h 2 u^2 + \frac g 2 h^2 . $$

First, lets write helper functions to compute the energy over the domain and then plot the time evolution of the energy for the Rusonov and the HLL solvers

```python
def computeEnergy(U,g):
    h = U[0]
    u = U[1]/U[0]
    return numpy.sum(h/2*u**2 + g/2*h**2)
def fvsolve2systemNoSlopeRec(riemann, U0, a=-1, b=1, n=20, tfinal=1, args = ()):
    h = (b - a)/n
    x = numpy.linspace(a+h/2, b-h/2, n) # Element midpoints (centroids)
    U0x = U0(x)
    Ushape = U0x.shape
    idxL = numpy.arange(-1, n-1)
    idxR = numpy.arange(1, n+1) % n
    def rhs(t, U):
        U = U.reshape(Ushape)
        jump = U[:,idxR] - U[:,idxL]
        r = numpy.zeros_like(jump)
        numpy.divide(U - U[:,idxL], jump, out=r, where=(jump!=0))
        g = jump / (2*h) #no slope limiter here!
        fluxL = riemann(U[:,idxL] + g[:,idxL] * h/2, U - g * h/2, *args)
        return (fluxL - fluxL[:,idxR]).flatten() / h
    hist = ode_rkexplicit(rhs, U0x.flatten(), h=h/5, tfinal=tfinal)
    return x, [(t, U.reshape(Ushape)) for t, U in hist]

def plotCompareEnergy(IC, limiter = limit_vl):
    x, hist = fvsolve2system(reimann_hll_shallowWater, IC, n=200, limit=limiter, args=(1,))
    x, hist2 = fvsolve2systemNoSlopeRec(reimann_hll_shallowWater, IC, n=200,args=(1,))

    E1 = []
    t1 = []
    for t, U in hist:
        E1.append(computeEnergy(U,1))
        t1.append(t)
    E2 = []
    t2 = []
    for t, U in hist2:
        E2.append(computeEnergy(U,1))
        t2.append(t)

    pyplot.figure()
    pyplot.plot(t1,E1)
    pyplot.plot(t2,E2)
    pyplot.legend({'with slope limiter', 'without slope limiter'})
    pyplot.ylabel('energy')
    pyplot.xlabel('t')
```

Next, let's look at the case of a supersonic shock:

```python
plotCompareEnergy(initial_shallowWater3)
pyplot.title('Supersonic Shock');
```

Looks like the height of the water goes negative at some point in the solver without a slope limiter. This is a serious drain on stability! The problem occurs when attempting to compute the shock speeds sL and sR where the term $\sqrt{g h}$ occurs. 

Next, let's look at a supersonic and a subsonic rarefaction:

```python
plotCompareEnergy(initial_shallowWater5)
pyplot.title('Supersonic Rarefaction')
plotCompareEnergy(initial_shallowWater4)
pyplot.title('Subsonic Rarefaction');
```

### 4: discussion

The same issue is happening in the for the supersonic rarefaction, though the subsonic rarefaction is captured without the slope limiter. After a point (though this time, farther forward in time), the height goes negative and breaks the solver. From what is computed, it appears that the solution without a slope limiter is slightly less diffusive (removes less energy from the solution overtime), as the computed energy in the eyeball norm is certainly greater for the solver sans slope limiter. We know that the HLL solver is slightly over diffusive in general, and it would appear in this simple numerical experiment that the slope limiter also adds some small amount of diffusion to the solution. If we were to use a full Reimann solver, we would recover some energy lost due to the HLL's artificial diffusion. A Rusonov solver would be more diffusive than the HLL solver, as it is even more conservative in capturing shock structures. Potentially, different limiters will have less numerical diffusion. Let's look at a couple:

```python
plotCompareEnergy(initial_shallowWater4, limiter = limit_bj)
pyplot.title('Subsonic Rarefaction, BJ Limiter');

plotCompareEnergy(initial_shallowWater4, limiter = limit_vl)
pyplot.title('Subsonic Rarefaction, VL Limiter');

plotCompareEnergy(initial_shallowWater4, limiter = limit_minmod)
pyplot.title('Subsonic Rarefaction, MinMod Limiter');
```

It would appear that the Barth-Jespersen incurs the least diffusion to the solution, while the MinMod limiter incurs the most. Clearly the choice of limiter can have a significant effect on the solution to this system.
